CREATE VENDOR PURCHASE SUMMARY TABLE

```python
vendor_purchase_summary = pd.read_sql("""
SELECT
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    pp.Volume,
    p.Description,
    SUM(p.Quantity) AS Quantity,
    AVG(p.PurchasePrice) AS PurchasePrice,
    SUM(p.Dollars) AS TotalPurchasePrice
FROM purchases p
JOIN purchase_prices pp
    ON p.Brand = pp.Brand
GROUP BY
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    pp.Volume,
    p.Description
""", conn)
```

```python
vendor_purchase_summary
```

```python
sales.columns
```

CREATE SALES SUMMARY TABLE 

```python
sales_summary = pd.read_sql(""" SELECT VendorNo,
Brand,
Description,
Volume,
SUM(SalesQuantity) AS TotalQuantity,
AVG(SalesPrice) AS SalesPrice,
SUM(SalesDollars) AS TotalRevenue
FROM sales
GROUP BY VendorNo, Brand, Description, Volume""",conn)
```

```python
sales_summary
```

```python
vendor_purchase_summary['Volume'] = pd.to_numeric(
    vendor_purchase_summary['Volume'], errors='coerce'
)

sales_summary['Volume'] = pd.to_numeric(
    sales_summary['Volume'], errors='coerce'
)
```

```python
vendor_sales_comparison = pd.merge(
    vendor_purchase_summary,
    sales_summary,
    left_on=['VendorNumber','Brand','Description','Volume'],
    right_on=['VendorNo','Brand','Description','Volume'],
    how='left'
)
```

```python
vendor_sales_comparison
```

```python
vendor_sales_comparison = vendor_sales_comparison.rename(
    columns={
        'Quantity': 'PurchaseQuantity',
        'TotalQuantity': 'SalesQuantity'
    }
)
vendor_sales_comparison
```

```python
vendor_sales_comparison.columns
```

```python
vendor_sales_comparison = vendor_sales_comparison.drop(columns=['VendorNo'])
```

```python
vendor_sales_comparison
```

```python
vendor_sales_comparison = vendor_sales_comparison[
    [
        'VendorNumber',
        'VendorName',
        'Brand',
        'Volume',
        'Description',
        'PurchaseQuantity',
        'PurchasePrice',
        'TotalPurchasePrice',
        'SalesQuantity',
        'SalesPrice',
        'TotalRevenue'
    ]
]
```

```python
vendor_sales_comparison
```

```python
freight_summary
```

```python
vendor_sales_comparison = vendor_sales_comparison.merge(
    freight_summary,
    on='VendorNumber',
    how='left'
)
```

```python
vendor_sales_comparison['Total_Freight'] = (
    vendor_sales_comparison['Total_Freight'].fillna(0)
)
```

```python
vendor_sales_comparison['NetRevenue'] = (
    vendor_sales_comparison['TotalRevenue']
    - vendor_sales_comparison['TotalPurchasePrice']
    - vendor_sales_comparison['Total_Freight']
)
vendor_sales_comparison['Freight_Percent'] = (
    vendor_sales_comparison['Total_Freight']
    / vendor_sales_comparison['TotalPurchasePrice']
)
```

```python
vendor_sales_comparison
```

```python
cols = ['SalesQuantity','SalesPrice','TotalRevenue']

vendor_sales_comparison[cols] = (
    vendor_sales_comparison[cols].fillna(0)
)
```
